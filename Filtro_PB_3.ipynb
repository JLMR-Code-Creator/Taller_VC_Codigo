```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

img_ruido = cv.imread('Fig0335.tif', cv.IMREAD_GRAYSCALE)

kernel_3x3 = np.ones((3,3), np.float32) / 9.0
#display(kernel_3x3)
```

```python
#from skimage.util import random_noise
#img_ruido = random_noise(img, mode='s&p')
#img_ruido = cv.normalize(img_ruido, None, 0, 255, cv.NORM_MINMAX, cv.CV_8U)

output3x3 = cv.filter2D(img_ruido, -1, kernel_3x3)
blurred = cv.GaussianBlur(img_ruido, (3, 3), cv.BORDER_DEFAULT)
mediana = cv.medianBlur(img_ruido, 3)

figure(figsize=(10, 8), dpi=200)
plt.subplot(1,4,1),plt.imshow(img_ruido, cmap='gray'), plt.title('Imagen original')
plt.xticks([]), plt.yticks([])
plt.subplot(1,4,2),plt.imshow(output3x3, cmap='gray'), plt.title('Salida filtro 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(1,4,3),plt.imshow(blurred, cmap='gray'), plt.title('Salida filtro gaussiano')
plt.xticks([]), plt.yticks([])
plt.subplot(1,4,4),plt.imshow(mediana, cmap='gray'), plt.title('Salida filtro mediana')
plt.xticks([]), plt.yticks([])
plt.show()
```
